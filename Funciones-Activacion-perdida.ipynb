```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoide

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-10,10,100)
```

```python
plt.plot(x,sigmoid(x))
plt.show()
```

Funcion de paso: es la funcion escalonada

```python
def step(x):
  return np.piecewise(x,[x<0.0,x>=0.0],[0,1])
```

```python
plt.plot(x,step(x))
plt.show()
```

# FUNCION DE PERDIDA

```python
def mse(y, y_hat, derivate=False):
  if derivate:
    return y_hat-y
  else:
    return np.mean((y_hat- y)**2)
```

```python
real=np.array([0,0,1,1])
prediccion=np.array([0.9,0.5,0.2,0.01])
```

```python
mse(real, prediccion)
```
